```python
# Estimators (number of trees)
data = [(0.76954,1900),(0.77088,1950),(0.77088,1975),(0.76819,2000),(0.76853,1925),(0.76987,1875),(0.76920,1850),(0.76752,1800),(0.76650,1000),(0.76853,3000)]
graph = pd.DataFrame(data,columns=['Accuracy on Subset','Number of Trees'])
plot = graph.plot.scatter(title = 'Change in accuracy based on change in number of Trees inputted',x='Number of Trees',y='Accuracy on Subset')
```

```python
# Class weights
data = {'Number of Trees' : [30,200,500,1000,2000,3000]}
data['Class Weights On'] = [0.64285,0.74561,0.76145,0.73820,0.76684,0.76482]
data['Class Weights Off'] = [0.63207,0.74932,0.76179,0.72641,0.76752,0.76415]
graph = pd.DataFrame(data)
graph.set_index('Number of Trees', inplace =True)
plot = graph.plot(title = 'How class weights change the accuracy of the classifier')
plot.set_ylabel('Accuracy on Subset')
```

```python
# Whole dataset vs 50/50 split vs 60/40 split ish at 3000 estimators
data = {'Dataset' : ['Whole dataset', '50/50 split', '43/57 split']}
data['Standardised Dataset'] = [0.40768,0.73146,0.76752]
data['Unmodified Dataset'] = [0.41812,0.67284,0.74022]
graph = pd.DataFrame(data)
graph.set_index('Dataset', inplace =True)
plot = graph.plot(title = 'How the training set changes the accuracy of the classifier')
plot.set_ylabel('Accuracy on Subset')
```
